```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly
import plotly.graph_objects as go
import datetime
```

```python
df = pd.read_csv('Datasets/city_temperature.csv',low_memory=False)
df.head(5)
```

```python
df = df.drop_duplicates()
df.isnull().sum()
```

```python
data = df[['Country','Year','AvgTemperature']].groupby(['Year','Country']).mean().sort_values('AvgTemperature').reset_index()
data = data.groupby('Year').apply(lambda group: group.iloc[:5]).reset_index(drop=True)
px.choropleth(data_frame=data,locationmode='country names',locations='Country',color='AvgTemperature',animation_frame='Year',title="Top five coldest countries over the years")
```

```python
df = pd.read_csv('Datasets/Environment_Temperature_change_E_All_Data_NOFLAG.csv', encoding='Latin-1')
```

```python
df = df[df['Element'] =='Temperature change']
df = df.fillna(0)
df=df.melt(id_vars=['Area Code', 'Area', 'Months Code', 'Months', 'Element Code', 'Element',
       'Unit'],var_name='Year', value_name='Temp Change')
df.head(5)
```

```python
df.drop(['Months Code','Element Code', 'Unit'], axis=1, inplace=True)
```

```python
df['Year'] = df['Year'].str[1:].astype('str')
df
```

```python
df['Months'].unique()
```

```python
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
       'August', 'September', 'October', 'November', 'December']
```

```python
df = df[df['Months'].isin(months)]
```

```python
df['Year'] = pd.to_datetime(df['Year'])
df = df.sort_values('Year',ascending=True)
df['Year'] = df['Year'].dt.strftime('%m/%d/%Y')
```

```python
top_10 = df.groupby('Area').sum().sort_values('Temp Change', ascending=False)[:10].reset_index()['Area']
bottom_10 = df.groupby('Area').sum().sort_values('Temp Change', ascending=True)[:10].reset_index()['Area']
```

```python
countries = top_10.append(bottom_10)
df = df[df['Area'].isin(countries)]
px.bar(df,x='Temp Change',y='Area',animation_frame='Year', hover_name='Temp Change',
      range_x=[-3.5,5.5], color='Area')
```
